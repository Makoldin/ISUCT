1. Получение ПСП


Дана строка, состоящая только из круглых скобок. Требуется определить, какое минимальное количество символов требуется удалить, чтобы она стала правильной скобочной последовательность.

Правильная скобочная последовательность (ПСП) — символьная последовательность, составленная в алфавите, состоящем из символов, сгруппированных в упорядоченные пары (типы скобок, графически обозначаемые «(» и «)», «[» и «]», «/*» и «*/» и т. п.), удовлетворяющая определённым правилам, обеспечивающим последовательную вложенность подпоследовательностей, обрамлённых открытой и закрытой скобкой одного типа.

Пример:

Ввод:    Вывод:
())(()        2

```python
def del_for_psp(value):
    open_val = 0
    closed_val = 0
    
    for val in value:
        if val == '(':
            open_val += 1  # открытая скобка
        else: 
            if open_val > 0:
                open_val -= 1 # скобка, которая закрывает открытую ранее
            else:
                closed_val += 1  # закрывающая скобка

    return closed_val + open_val

v = del_for_psp('())(()')
v
```

2. Ближайший меньший

Задан массив чисел, требуется для каждого числа найти ближайшее справа строго меньшее число.

На вход передается две строки: в первой N - количество элементов в массиве и во второй N чисел

Пример:

Ввод:                              Вывод:
10                                   -1 4 3 4 -1 6 9 8 9 -1 

1 2 3 2 1 4 2 5 3 1


```python
n = 10
arr = [1, 2, 3, 2, 1, 4, 2, 5, 3, 1]

result = []

for i in range(n):
    found_index = -1

    for j in range(i + 1, n):
        if arr[j] < arr[i]: #ближайшее справа строго меньшее число
            found_index = j
            break  # нашли ближайшее
    
    result.append(found_index) #добавляем в result

result
```
